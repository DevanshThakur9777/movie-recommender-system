```python
movies.dropna(inplace=True)
movies.shape
```

```python
def convert(obj):
    return [i['name'] for i in ast.literal_eval(obj)]
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
def convert3(obj):
    obj = ast.literal_eval(obj)
    return [i['name'] for i in obj[:3]]
```

```python
movies['cast'] = movies['cast_x'].apply(convert3)
```

```python
movies.head(2)
```

```python
def fetch_director(obj):
    obj = ast.literal_eval(obj)

    for i in obj:
        if i['job'] == 'Director':
            return [i['name']]
    return []
```

```python
movies['crew'] = movies['crew_x'].apply(fetch_director)
```

```python
movies.head(1)
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())  #convert string to list
```

```python
movies['overview'][0]
```

```python
# space remove kar rahe hain taki words combine ho jaye (better matching ke liye)
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
# sab columns ko combine karke ek single tags column bana rahe hain
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies[['title','tags']].head(1)
```

```python
movies.head(2)
```

```python
new_df = movies[['id','title','tags']]
```

```python
new_df
```

```python
# list ko string me convert kar rahe hain (vectorization ke liye)
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head(2)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head(1)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# words ko root form me convert kar rahe hain (similar words match karne ke liye)
def stem(text):
    return " ".join([ps.stem(i) for i in text.split()])
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head(1)
```

```python
# TF-IDF use kar rahe hain (better accuracy ke liye)
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=5000, stop_words='english')
vectors = tfidf.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
# cosine similarity calculate kar rahe hain
# ye batata hai ki ek movie dusri movie se kitni similar hai
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
#ye function similar movies suggest karega
def recommend(movie):
  index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[index]
  movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])
  recommended_movies = []
  for i in movies_list[1:6]:
        recommended_movies.append(new_df.iloc[i[0]].title)
  return recommended_movies
```

```python
recommended = recommend('Batman Begins')
recommended
```

```python
#correct movie list
relevant_movies = [
    "The Dark Knight",
    "The Dark Knight Rises",
    "Batman v Superman: Dawn of Justice",
    "Man of Steel"
]
```

```python
def precision_at_k(recommended, relevant, k=5):
    # top k recommended movies select kar rahe hain
    recommended = recommended[:k]

    # correct matches count karne ke liye variable
    count = 0

    # har recommended movie ko check kar rahe hain
    for movie in recommended:
        if movie in relevant:
            count += 1

    # precision = correct predictions / total predictions
    return count / k
```

```python
score = precision_at_k(recommended,relevant_movies,5)
score
```
